# Step 5 -Statistical Test:

#### Mann Whitney Test

```python
#Attrition Vs Distance from Home
from scipy.stats import mannwhitneyu

employees_attrition_yes = employees_dataset[employees_dataset["Attrition"] == 'Yes']
employees_attrition_yes
```

```python
employees_attrition_no = employees_dataset[employees_dataset["Attrition"] == 'No']
employees_attrition_no
```

#### Mann Whitney U Test:  Attrition Vs Distance From Home 

```python
stats,p = mannwhitneyu(employees_attrition_yes['DistanceFromHome'],employees_attrition_no['DistanceFromHome'])
print("r =",stats,"p =",p)

print("\nH0: There is no significant differences between Distance From Home of current employees and ex-employees")
print("Ha: There is significant differences between Distance From Home of current employees and ex-employees")
if(p<0.05):
    print("\np<0.05")
    print("\nH0 is rejected and Ha is accepted. \nThere is significant difference between DistanceFromHome of current employees and ex-employees.")
else:
    print("\np>0.05")
    print("\nResult: H0 is Accepted, There is no significant difference between DistanceFromHome of current employees and ex-employees.")
```
