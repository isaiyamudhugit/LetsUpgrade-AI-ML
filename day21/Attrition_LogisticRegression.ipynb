# Assignment - Logistic Regression - Attrition

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
dataset = pd.read_csv("general_data.csv")
```

```python
dataset.head()
```

```python
dataset.dropna()
dataset.drop_duplicates()
```

```python
label_encoder = preprocessing.LabelEncoder()
dataset['BusinessTravel'] = label_encoder.fit_transform(dataset['BusinessTravel'])
dataset['Department'] = label_encoder.fit_transform(dataset['Department'])
dataset['EducationField'] = label_encoder.fit_transform(dataset['EducationField'])
dataset['Gender'] = label_encoder.fit_transform(dataset['Gender'])
dataset['JobRole'] = label_encoder.fit_transform(dataset['JobRole'])
dataset['MaritalStatus'] = label_encoder.fit_transform(dataset['MaritalStatus'])
dataset['Attrition'] = label_encoder.fit_transform(dataset['Attrition'])
```

```python
import statsmodels.api as sm
dataset.columns
```

```python
Y = dataset['Attrition']
dataset1 = dataset.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'],axis=1)
dataset1['NumCompaniesWorked']= np.where(dataset1['NumCompaniesWorked'].isnull(), dataset1['NumCompaniesWorked'].mean(), dataset1['NumCompaniesWorked'] )
dataset1['TotalWorkingYears']= np.where(dataset1['TotalWorkingYears'].isnull(), dataset1['TotalWorkingYears'].mean(), dataset1['TotalWorkingYears'] )
dataset1['NumCompaniesWorked'] = np.round(dataset1['NumCompaniesWorked'])
dataset1['TotalWorkingYears'] = np.round(dataset1['TotalWorkingYears'])
dataset1.head()
```

```python
X = dataset1[['Age','BusinessTravel','Department','DistanceFromHome','Education','EducationField','Gender','JobLevel','JobRole','MaritalStatus','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
X1 = sm.add_constant(X)
```

```python
Logistic = sm.Logit(Y,X1)
```

```python
result = Logistic.fit()
```

```python
result.summary()
```
