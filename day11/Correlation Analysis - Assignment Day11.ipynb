```python
from sklearn import preprocessing
```

```python
from scipy.stats import pearsonr 
```

```python
# 1 Correlation between Attrition and Age:

stats,p = pearsonr(employee_dataset['Attrition'],employee_dataset['Age'])

print(stats,p)

if(p<0.05):
    print("There is significant correlation between Attrition and Age. Rejecting Null Hypothesis and Accepting the Alternate Hypothesis")
else:
    print("There is no significant correlation between Attrition and Age. Accepting Null Hypothesis")

#
```

### Correlation between Attrition and Age:

 Here,
    r = -0.15920500686577957 
    
    So,  there is low negative correlation between Attrition and Age 
    
    p = 1.996801615887198e-26 
        
    P value is < 0.05, Rejecting H0 and Accepting H1, hence there is significant correlation between Attrition & Age
